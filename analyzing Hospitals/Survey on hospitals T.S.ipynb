```python
import pandas as pd
import numpy as np
Hospitals=pd.read_csv('Hospitals.csv')
```

# DATA SET OF HOSPITALS IN T.S

```python
Hospitals
```

 # PIE GRAPH OF DISTRICT vs Health_Sub-Centres

```python
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']=(55,25)
```

```python
plt.bar(Hospitals.loc[:,"District"], Hospitals.loc[:,"Health_Sub-Centres"])
plt.xlabel('District') 
plt.ylabel('no of. Health_Sub-Centres') 
```

 # PIE CHART OF  Primary_Health_Centres IN DISTRICTS

```python
plt.rcParams['figure.figsize']=(25,10)
plt.pie(x=Hospitals.loc[:,"Primary_Health_Centres"] ,labels=Hospitals.loc[:,"District"],radius=1.5,autopct='%1.1f%%')
```

# GRAPH OF DISTRICT vs Community_Health_Centres	

```python
plt.rcParams['figure.figsize']=(55,25)

plt.plot(Hospitals.loc[:,"District"],Hospitals.loc[:,"Community_Health_Centres"],color='blue', marker='o', linestyle='dashed',linewidth=2, markersize=12)

plt.show()
```

# GRAPH OF DISTRICT vs Ayurveda_Hospitals_(incl.Dispensaries)	

```python
plt.rcParams['figure.figsize']=(55,25)
plt.bar(Hospitals.loc[:,"District"], Hospitals.loc[:,"Ayurveda_Hospitals_(incl.Dispensaries)"],width=1)
```

# GRAPH OF DISTRICT vs Homeopathic Hospitals (incl. Dispensaries)	

```python
plt.rcParams['figure.figsize']=(55,25)
plt.plot(Hospitals.loc[:,"District"],Hospitals.loc[:,"Homeopathic Hospitals (incl. Dispensaries)"],color='blue', marker='o', linestyle='dashed',linewidth=2, markersize=12)
plt.plot(Hospitals.loc[:,"District"],Hospitals.loc[:,"Unani Hospitals (incl. Dispensaries)"],color='black', marker='o', linestyle='dashed',linewidth=2, markersize=12)
plt.plot(Hospitals.loc[:,"District"],Hospitals.loc[:,"Naturopathy Hospitals (incl. Dispensaries)"],color='red', marker='o', linestyle='dashed',linewidth=2, markersize=12)
plt.legend(["Homeopathic Hospitals","Unani Hospitals","Naturopathy Hospitals"])
plt.show()
```
